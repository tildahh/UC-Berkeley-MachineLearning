
# Introduction to Data Visualization with Pandas

## Overview

## Dataset Description
The dataset we're using is sourced from GeeksforGeeks and contains various details about NBA players. It includes attributes like player names, ages, teams, games played, and various performance metrics. This rich dataset provides an excellent opportunity to apply and showcase different types of data visualizations.

## Objectives
In this notebook, we will:
- Learn how to load and inspect the NBA dataset using pandas.
- Explore different types of plots (histograms, scatter plots, line plots, and bar charts) to visualize various aspects of the NBA data.
- Understand how to customize these plots to make them more informative and visually appealing.

By the end of this notebook, you will gain practical experience in using pandas for data visualization and draw insights from the NBA dataset.

Let's begin our journey into data visualization with pandas, using this interesting NBA dataset!


```python
# Reading data into a pandas DataFrame
# This cell performs data manipulation and visualization tasks
# import data from pandas
import pandas as pd
import matplotlib.ticker as ticker
import os

#Make directory called plots
os.makedirs("plots", exist_ok=True)

#create the DataFrame utilizing Pandas 
data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 

#generate all info for this dataset
data.info()

#Remove the last line of NAN
data = data.iloc[:-1 , :]

#Ignore any warnings
import warnings
warnings.filterwarnings('ignore')

data.head(10)
```

###  Part One: In this exercise, generate a basic histogram.


This part focuses on a specific type of chart, illustrating the versatility of pandas in handling different data visualization needs.

```python
#Please generate a basic histogram and save the histogram as a png file

ax = data.plot(kind = 'hist', y=['Number','Age','Weight','Salary'], figsize=(10, 6))
ax.set_xlabel('Value')
ax.legend(title='Key')

ax.figure.savefig('histogram.png')
```

###  Part Two: In this exercise, generate a basic scatter plot. 


This part focuses on a specific type of chart, illustrating the versatility of pandas in handling different data visualization needs.

```python
#Please generate a basic scatter plot utilizing Age as the x axis and Salary as the y axis

scat = data.plot(kind='scatter', x='Age', y='Salary', grid=True, figsize=(10, 6), title='Age vs. Salary')
scat.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}')) 
print(scat)

scat.figure.savefig('scatterPlot.png')
```

###  Part Three: In this exercise, generate a basic lineplot. 

This part focuses on a specific type of chart, illustrating the versatility of pandas in handling different data visualization needs.

```python
# Please generate a basic line plot by name as the x axis and salary as the y axis 
#for the "Golden State Warriors" using the nba data set. 

teamdata = data.filter(items=['Team', 'Name', 'Salary'])
teamdata = teamdata.loc[data.Team == 'Golden State Warriors', 'Name':'Salary']
teamdata.set_index('Name')

lines = teamdata.plot(kind='line', x='Name', y='Salary',grid=True, title='Golden State Warriors Salary by Player')
lines.set_xticklabels(lines.get_xticklabels(), rotation=90)
lines.set_xlabel('Player Name')
lines.set_ylabel('Salary')
print(lines)
lines.figure.savefig('linePlot.png')
```

###  Part Four: In this exercise, generate a basic bar chart. 

This part focuses on a specific type of chart, illustrating the versatility of pandas in handling different data visualization needs.

```python
# Please generate a basic bar plot by name as the x axis and salary as the y axis of the teamdata dataframe.
 
bar = teamdata.plot(kind='bar', x='Name', y='Salary', grid=True, title='Golden State Warriors Salary by Player')
bar.set_ylabel('Salary')

print(bar)
bar.figure.savefig('barChart.png')
```
